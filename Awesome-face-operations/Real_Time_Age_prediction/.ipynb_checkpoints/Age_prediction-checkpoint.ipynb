```python
import numpy as np
import pandas as pd 
import cv2
from IPython.display import Image
import matplotlib.pyplot as plt
import os
import tensorflow as tf
from tensorflow.keras import layers, models
from sklearn.model_selection import train_test_split
print(os.listdir("../input/utkface-new/"))
```

```python
def imshow(img):
    plt.imshow(img, interpolation='bicubic')
    plt.xticks([]), plt.yticks([]) 
    plt.show()
tf.__version__
```

```python
onlyfiles = os.listdir("../input/utkface-new/UTKFace/")
y = np.array([[i.split('_')[0]] for i in onlyfiles])
# y = np.array([[i.split('_')[1] for i in onlyfiles]]).T
print(y.shape)
print(y[0])
```

```python
X_data =[]
for file in onlyfiles:
    face = cv2.imread("../input/utkface-new/UTKFace/"+file,cv2.IMREAD_COLOR)
    face = cv2.cvtColor(face, cv2.COLOR_BGR2RGB)
    face =cv2.resize(face, (32,32) )
    X_data.append(face)
X_data=np.array(X_data)
X_data.shape
```

```python
X = np.squeeze(X_data)
imshow(X[1])
print(y[1])
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.33)
```

```python
y_train.shape
```

```python
y_train = y_train[:,0].astype(int)
y_valid = y_valid[:,0].astype(int)
```

```python
y_train
```

```python
def gen_model():
    model = models.Sequential()
    model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)))
    model.add(layers.Conv2D(32, (3, 3), activation='relu'))
    model.add(layers.MaxPooling2D((2, 2)))
    model.add(layers.Dropout(0.3))
    model.add(layers.Conv2D(64, (3, 3), activation='relu'))
    model.add(layers.Conv2D(64, (3, 3), activation='relu'))
    model.add(layers.MaxPooling2D((2, 2)))
    model.add(layers.Dropout(0.3))
    model.add(layers.Conv2D(84, (3, 3), activation='relu'))
    model.add(layers.Dropout(0.3))
    model.add(layers.Flatten())
    model.add(layers.Dense(64,activation='relu'))
    model.add(layers.Dense(1,activation='relu',name='age_out'))
    model.compile(optimizer='Adam',
                  loss=['mse']
                  )
    tf.keras.utils.plot_model(model, 'model.png',show_shapes=True)
    return model
model=gen_model()
Image('model.png')
```

```python
print(X_train.shape)
y_train.shape
```

```python
import random
random_id=random.random()
model.summary()
callbacks = [
    tf.keras.callbacks.EarlyStopping(patience=75, monitor='val_loss',restore_best_weights=True),
    tf.keras.callbacks.TensorBoard(log_dir='./logs/'+str(random_id))
]
history=model.fit(X_train, y_train, epochs=400,batch_size=240,validation_data=(X_valid,y_valid),callbacks=callbacks, shuffle=True)
```

```python
model.evaluate(X_valid,y_valid)
```

```python
p_id=1782
imshow(X_valid[p_id])
print(y_valid[p_id])
#print(y_valid[0][p_id],y_valid[1][p_id])
print(model.predict(np.expand_dims(X_valid[p_id],axis=0)))
#np.expand_dims(X_valid[p_id],axis=0)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.show()
```
